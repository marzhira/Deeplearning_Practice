```python
#보스턴 집값예측
import pandas as pd
import tensorflow as tf
```

```python
boston = pd.read_csv('https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv')
print(boston.columns)
boston
```

```python
boston_x = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
boston_y = boston[['medv']]
print(boston_x.shape, boston_y.shape)
```

```python
#모델 생성
x = tf.keras.layers.Input(shape=[13])
y = tf.keras.layers.Dense(1)(x)
model = tf.keras.models.Model(x, y)
model.compile(loss='mse')
```

```python
#학습
model.fit(boston_x, boston_y, epochs=500)
```

```python
model.predict(boston_x[0:5])
```

```python
#모델의 수식 확인
model.get_weights()
```
