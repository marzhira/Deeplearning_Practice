```python
import pandas as pd
import tensorflow as tf
```

```python
 #파일 불러오기
 iris = pd.read_csv('https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv')
 print(iris.columns)
 iris.head()
```

```python
#원핫인코딩
iris_cleaned = pd.get_dummies(iris)
iris_cleaned
```

```python
#독립 종속변수 분류
iris_x = iris_cleaned[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
iris_y = iris_cleaned[['품종_setosa', '품종_versicolor', '품종_virginica']]
iris_x.head()
```

```python
#모델의 구조 만들기
x = tf.keras.layers.Input(shape=[4])
y = tf.keras.layers.Dense(3, activation='softmax')(x)
model = tf.keras.models.Model(x, y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```python
#모델 학습
model.fit(iris_x, iris_y, epochs=1000)
```

```python
#모델 이용
model.predict(iris_x[0:5])
```
